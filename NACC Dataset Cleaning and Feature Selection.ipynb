```python
i=0
rows_list = []

for index, row in dropp.iterrows():
    naccid = row['NACCID']
    if (naccid == 'NACC000385'):
        print('we are a go')
    if dropees.__contains__(naccid):
        print('dropped')
        i+=1
    else:
        dict1 = {}
        dict1.update(row)
        rows_list.append(dict1)
    print(i)
```
